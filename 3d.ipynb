```python
import trimesh

mesh = trimesh.load_mesh('disc.obj')
vertices = mesh.vertices
faces = mesh.faces
```

```python
import pandas as pd

# Prepare vertex data
vertex_data = pd.DataFrame(vertices, columns=['x', 'y', 'z'])

# Prepare face data
face_colors = ['rgb(255,0,0)'] * len(faces)  # Example color
face_data = pd.DataFrame(faces, columns=['i', 'j', 'k'])
face_data['facecolor'] = face_colors

# Save to CSV
vertex_data.to_csv('vertices.csv', index=False)
face_data.to_csv('faces.csv', index=False)
```

```python
import plotly.graph_objects as go
import pandas as pd

# Load the data
vertices = pd.read_csv('vertices.csv')
faces = pd.read_csv('faces.csv')


# Create the plot
fig = go.Figure(data=[go.Mesh3d(
    x=vertices['x'],
    y=vertices['y'],
    z=vertices['z'],
    i=faces['i'],
    j=faces['j'],
    k=faces['k'],
    facecolor=faces['facecolor'],
    opacity=0.5  # Optional: Adjust opacity for better visualization
)])


# Update layout with axis ranges
fig.update_layout(
    scene=dict(
        aspectmode='data'  # Ensures aspect ratio is proportional
    ),
)
fig.show()
```
